```python
import numpy as np
atleti = [
    "Marco",
    "Luca",
    "Sara",
    "Paolo",
    "Anna",
    "Giulia"
]

altezza = [182, 175, 168, 160, 185, 165]
peso = [82, 76, 60, 55, 88, 58]
torace = [102, 98, 85, 80, 106, 83]
coscia = [61, 58, 50, 48, 65, 49]
imc = [24.8, 24.8, 21.3, 21.5, 25.7, 21.3]



# calcolare la media di una variabile
def calcola_media(lista):
    return sum(lista) / len(lista)


# calcolare la deviazion standard campionaria

def calcola_deviazione_standard(lista, media):
    varianza = sum((x - media) ** 2 for x in lista) / (len(lista) - 1)
    return varianza ** 0.5

# standardizza le variabili tarmite z-score 

def standardizza(lista):
    media = calcola_media(lista)
    dev_std = calcola_deviazione_standard(lista, media)
    return [(x - media) / dev_std for x in lista]

# calcolo covarianza

def calcola_covarianza(x, y):
    n = len(x)
    media_x = calcola_media(x)
    media_y = calcola_media(y)
    
    somma_prodotti = 0
    for i in range(n):
        somma_prodotti += (x[i] - media_x) * (y[i] - media_y)
        
    return somma_prodotti / (n - 1)


# 3. CREAZIONE DATI STANDARDIZZATI (Z-SCORE)
altezza_z = standardizza(altezza)
peso_z = standardizza(peso)
torace_z = standardizza(torace)
coscia_z = standardizza(coscia)
imc_z = standardizza(imc)

dati_z = [
    altezza_z,
    peso_z, 
    torace_z,
    coscia_z, 
    imc_z 
]

# Nomi delle variabili (con le virgole!)
nomi_variabili = [
    "altezza",
    "peso",
    "torace",
    "coscia",
    "imc"
]

n_var = len(dati_z)


# creazione matrice di covarianza

matrice_cov = []

for i in range(n_var):
    riga = []
    for j in range(n_var):
        cov = calcola_covarianza(dati_z[i], dati_z[j])
        riga.append(round(cov, 4))
    matrice_cov.append(riga)

# calcolare utovalori e autovettori

cov_np = np.array(matrice_cov)
print(cov_np)

# Autovalori e Autovettori

autovalori, autovettori = np.linalg.eig(cov_np)
print("AutoValori\n")

for i, valore in enumerate(autovalori):
    print(f"PC{i+1}: {valore:.2f}")
    

print("AutoVettori\n")

for i in range(len(autovalori)):
    print(f"\nPC{i+1}")

    for nome, coeff in zip(nomi_variabili, autovettori[:, i]):
        print(f"{nome:10} {coeff:.2f}")


```
