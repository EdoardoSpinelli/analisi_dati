```python
import numpy as np
atleti = [
    "Marco",
    "Luca",
    "Sara",
    "Paolo",
    "Anna",
    "Giulia"
]

altezza = [182, 175, 168, 160, 185, 165]
peso = [82, 76, 60, 55, 88, 58]
torace = [102, 98, 85, 80, 106, 83]
coscia = [61, 58, 50, 48, 65, 49]
imc = [24.8, 24.8, 21.3, 21.5, 25.7, 21.3]



# calcolare la media di una variabile
def calcola_media(lista):
    return sum(lista) / len(lista)


# calcolare la deviazion standard campionaria

def calcola_deviazione_standard(lista, media):
    varianza = sum((x - media) ** 2 for x in lista) / (len(lista) - 1)
    return varianza ** 0.5

# standardizza le variabili tarmite z-score 

def standardizza(lista):
    media = calcola_media(lista)
    dev_std = calcola_deviazione_standard(lista, media)
    return [(x - media) / dev_std for x in lista]

# calcolo covarianza

def calcola_covarianza(x, y):
    n = len(x)
    media_x = calcola_media(x)
    media_y = calcola_media(y)
    
    somma_prodotti = 0
    for i in range(n):
        somma_prodotti += (x[i] - media_x) * (y[i] - media_y)
        
    return somma_prodotti / (n - 1)


# 3. CREAZIONE DATI STANDARDIZZATI (Z-SCORE)
altezza_z = standardizza(altezza)
peso_z = standardizza(peso)
torace_z = standardizza(torace)
coscia_z = standardizza(coscia)
imc_z = standardizza(imc)

dati_z = [
    altezza_z,
    peso_z, 
    torace_z,
    coscia_z, 
    imc_z 
]

# Nomi delle variabili (con le virgole!)
nomi_variabili = [
    "altezza",
    "peso",
    "torace",
    "coscia",
    "imc"
]

n_var = len(dati_z)


# creazione matrice di covarianza

matrice_cov = []

for i in range(n_var):
    riga = []
    for j in range(n_var):
        cov = calcola_covarianza(dati_z[i], dati_z[j])
        riga.append(round(cov, 4))
    matrice_cov.append(riga)

# calcolare utovalori e autovettori

cov_np = np.array(matrice_cov)
print(cov_np)

# Autovalori e Autovettori

autovalori, autovettori = np.linalg.eig(cov_np)
print("AutoValori\n")

for i, valore in enumerate(autovalori):
    print(f"PC{i+1}: {valore:.2f}")
    

print("AutoVettori\n")

for i in range(len(autovalori)):
    print(f"\nPC{i+1}")

    for nome, coeff in zip(nomi_variabili, autovettori[:, i]):
        print(f"{nome:10} {coeff:.2f}")


```

## Esercizio rifatto

```python
import numpy as np

aziende = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "L"]

fatturato   = [1000, 200, 800, 500, 300, 700, 400, 600, 900, 100]
esg         = [  30,  85,  90,  20,  70,  40,  95,  25,  60,  80]
innovazione = [  80,  20,  50,  90,  40,  70,  10,  60,  30,  85]
turnover    = [  35,   5,  15,  30,  10,  25,  20,   8,  28,  12]
nps         = [  20,  80,  40,  10,  90,  30,  70,  50,  60,  95]
sicurezza   = [  90,  60,  20,  85,  55,  15,  75,  40,  30,  70]


# 1. FUNZIONI STATISTICHE DI BASE
def calcola_media(lista):
    return sum(lista) / len(lista)

def calcola_deviazione_standard(lista, media):
    varianza = sum((x - media) ** 2 for x in lista) / (len(lista) - 1)
    return varianza ** 0.5

def standardizza(lista):
    media = calcola_media(lista)
    dev_std = calcola_deviazione_standard(lista, media)
    return [(x - media) / dev_std for x in lista]

def calcola_covarianza(x, y):
    n = len(x)
    media_x = calcola_media(x)
    media_y = calcola_media(y)
    somma_prodotti = sum((x[i] - media_x) * (y[i] - media_y) for i in range(n))
    return somma_prodotti / (n - 1)


# 2. DIZIONARI PER COMODITÀ (Salvano medie e dev_std utili per l'Azienda M)
dati_grezzi = {
    "fatturato": fatturato,
    "esg": esg,
    "innovazione": innovazione,
    "turnover": turnover,
    "nps": nps,
    "sicurezza": sicurezza
}

nomi_variabili = list(dati_grezzi.keys())
medie = {k: calcola_media(v) for k, v in dati_grezzi.items()}
dev_std = {k: calcola_deviazione_standard(v, medie[k]) for k, v in dati_grezzi.items()}

# Creazione dataset standardizzato
dati_z = [standardizza(dati_grezzi[nome]) for nome in nomi_variabili]
n_var = len(dati_z)


# 3. MATRICE DI COVARIANZA
matrice_cov = []
for i in range(n_var):
    riga = []
    for j in range(n_var):
        cov = calcola_covarianza(dati_z[i], dati_z[j])
        riga.append(round(cov, 4))
    matrice_cov.append(riga)

cov_np = np.array(matrice_cov)


# 4. AUTOVALORI, AUTOVETTORI E VARIANZA SPIEGATA
autovalori, autovettori = np.linalg.eig(cov_np)

print("--- RISULTATI PCA ---")
varianza_totale = sum(autovalori)
for i, valore in enumerate(autovalori):
    var_spiegata = (valore / varianza_totale) * 100
    print(f"PC{i+1}: Autovalore = {valore:.2f} | Varianza Spiegata = {var_spiegata:.2f}%")


print("\n--- COEFFICIENTI AUTOVETTORI (LOADINGS) ---")
for i in range(len(autovalori)):
    print(f"\nPC{i+1}:")
    for j, nome in enumerate(nomi_variabili):
        print(f"  {nome:12}: {autovettori[j, i]:.2f}")


# 5. CASO AZIENDA M
print("\n--- PROIEZIONE AZIENDA M ---")
azienda_m_grezza = {
    "fatturato": 600,
    "esg": 80,
    "innovazione": 75,
    "turnover": 12,
    "nps": 85,
    "sicurezza": 70
}

# Standardizzazione di M con i parametri storici (medie e dev_std del cluster iniziale)
azienda_m_z = []
for nome in nomi_variabili:
    z_score = (azienda_m_grezza[nome] - medie[nome]) / dev_std[nome]
    azienda_m_z.append(z_score)

azienda_m_z = np.array(azienda_m_z)

# Calcolo del punteggio (Score) di M sulle Componenti Principali
for i in range(len(autovalori)):
    score_m = np.dot(azienda_m_z, autovettori[:, i])
    print(f"Punteggio di Azienda M su PC{i+1}: {score_m:.2f}")
```
