## Importazione delle librerie e creazione del dataset del titnaic 

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carica il dataset del Titanic integrato in seaborn
df = sns.load_dataset("titanic")
print(df.head())
````

## 1. Qual è la distribuzione dell'età dei passeggeri?

Per vedere come si distribuisce un'unica variabile numerica (l'età), il grafico ideale è l'istogramma (con o senza la linea della densità KDE).

````python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='age', kde=True, bins=30, color='skyblue')
plt.title("Distribuzione dell'età dei passeggeri")
plt.xlabel("Età")
plt.ylabel("Numero di Passeggeri")
plt.show()
````

## 2. Quanti passeggeri ci sono per classe?

Qui abbiamo una variabile categoriale (class o pclass). Per contare quanti elementi ci sono in ogni categoria, usiamo un count plot.

````python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='class', palette='Set2')
plt.title("Numero di passeggeri per classe")
plt.xlabel("Classe")
plt.ylabel("Conteggio")
plt.show()
````

## 3. C'è una relazione tra età e tariffa pagata?
Per mettere in relazione due variabili numeriche (età vs tariffa/fare), il grafico perfetto è lo scatter plot (grafico a dispersione).

````python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='fare', alpha=0.6, hue='survived')
plt.title("Relazione tra Età e Tariffa Pagata")
plt.xlabel("Età")
plt.ylabel("Tariffa (Fare)")
plt.show()
````

## 4. Le tariffe cambiano in base alla classe?
Per confrontare una variabile numerica (tariffa) divisa per categorie (classe), il boxplot (grafico a scatola e baffi) o il barplot sono ottimi. Il boxplot ti mostra anche la distribuzione e i valori fuori norma.

````python
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, x='class', y='fare', palette='Pastel1')
plt.title("Tariffe in base alla Classe")
plt.xlabel("Classe")
plt.ylabel("Tariffa (Fare)")
plt.show()
````

## 5. Ci sono valori anomali o gruppi molto diversi tra loro?
Guardando i grafici precedenti (soprattutto lo scatter plot e il boxplot della tariffa), noterai subito dei valori anomali (outliers). Ad esempio:

C'è un piccolo gruppo di passeggeri in prima classe che ha pagato una tariffa altissima (oltre 500 dollari), totalmente staccata dal resto.

Puoi anche usare un pairplot per avere una panoramica visiva immediata di tutte le relazioni e scovare raggruppamenti strani:

````python
# Un riepilogo visivo combinato per trovare anomalie o gruppi
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='Dark2')
plt.show()
````
