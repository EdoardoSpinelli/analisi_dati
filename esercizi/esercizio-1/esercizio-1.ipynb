```python
import pandas as pd
import numpy as np

bevande_alcoliche_vendute = [
    12, 15, 18, 20, 14, 16, 22, 19, 17, 13, 
    21, 24, 26, 18, 16, 15, 23, 25, 14, 20
]


fatturato = [
    120.5, 150.2, 180.0, 210.3, 140.8, 160.4, 
    220.7, 190.6, 170.1, 130.9, 215.4, 240.8, 
    260.5, 185.2, 165.7, 155.3, 230.9, 250.1, 145.6, 205.0
]


eta = [
    22, 25, 31, 28, 35, 40, 27, 29, 33, 24,
    30, 38, 42, 26, 34, 36, 41, 39, 23, 32
]


sesso = [
    "F", "M", "F", "M", "F", "M", "F", "F", "M", "F",
    "M", "M", "F", "M", "F", "F", "M", "F", "M", "F"
]



df = pd.DataFrame({
    "bevande_alcoliche_vendute": bevande_alcoliche_vendute,
    "fatturato": fatturato,
    "eta": eta,
    "sesso": sesso
})


print(df)


print("--- 1. INDICATORI STATISTICI ---")

stats = df.describe().round(2)



q1 = df.quantile(0.25, numeric_only=True)
q3 = df.quantile(0.75, numeric_only=True)
stats.loc['IQR'] = (q3 - q1).round(2)

stats = stats.rename(index={'50%': 'median'})

print(stats.loc[['mean', 'median', '25%', '75%', 'IQR', 'std']])
print("\n" + "-"*50 + "\n")


print("--- 2. CONTEGGIO SESSO (Value Counts) ---")


print(df["sesso"].value_counts())
print("\n" + "-"*50 + "\n")


print("--- 3. MATRICE DI CORRELAZIONE ---")

df_numerico = df[["bevande_alcoliche_vendute", "fatturato", "eta"]]
print(df_numerico.corr().round(2))
```
