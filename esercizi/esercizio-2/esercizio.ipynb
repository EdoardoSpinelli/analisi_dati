## Passo 3: Cacola la mediana si stipendio

```python
# Passo 3: Calcola mediana di stipendio escludendo None

nomi = ["Luca", "Sara", "Marco", "Paolo", "Anna"]
eta = [25, None, 35, 310, 29]
stipendio = [2500, 2800, None, 3200, 2900]
reparto = ["IT", "HR", "Marketing", "IT", "HR"]

def calcola_mediana_stipendio(lista_stipendio):
    """
    Calcola la mediana di stipendio escludendo None
    e sostituisce il valore mancante con la mediana
    """
    
    stipendio_valido = [s for s in lista_stipendio if s is not None]

    # Ordino i valori
    stipendio_valido.sort()

    # calcolo della mediana
    n = len(stipendio_valido)
    if n % 2 == 1:  # dispari
        mediana = stipendio_valido[n // 2]
    else: # pari
        mediana = (stipendio_valido[n // 2 - 1] + stipendio_valido[n // 2]) / 2

    print(f"La mediana dello stipendio escludendo il None è: {mediana}")
    
    # Sostituisci None con la mediana
    stipendio_corretto = [mediana if s is None else s for s in lista_stipendio]
    print(f"Stipendio dopo la sostituzione: {stipendio_corretto}")
    
    return stipendio_corretto

# Applica la funzione
stipendio = calcola_mediana_stipendio(stipendio)
```

### Obiettivo:

- correlazione tra ore di studio a settimana, 
- ore di sonno a notte 
- e voto medio esami 

* Calcoli la correlazione di Person tra tuute le coppie di variabili
* stampi una matrice di correlazione ben formattata
* interpetri a parola alemno tre correlazioni
* con una soglia di correlazione di 0.95 analizza la rindondanza

```python
ore_studio = [5, 8, 3, 10, 7, 6, 9, 4, 2, 11]
ore_sonno  = [7, 6, 8, 5, 6, 7, 5, 8, 9, 4]
voto_medio = [24, 27, 22, 30, 26, 25, 29, 23, 21, 31]

variabili = {
    "ore_studio":           ore_studio,
    "ore_sonno":            ore_sonno,
    "voto_medio":           voto_medio
}


nomi_var = list(variabili.keys())

def calcola_media(lista):
    return sum(lista) / len(lista)

def calcola_correlazione(x, y):
    n = len(x)
    media_x = calcola_media(x)
    media_y = calcola_media(y)

    numeratore = sum((x[i] - media_x) * (y[i] - media_y) for i in range(n))
    denom_x    = sum((x[i] - media_x) ** 2 for i in range(n)) ** 0.5
    denom_y    = sum((y[i] - media_y) ** 2 for i in range(n)) ** 0.5

    if denom_x == 0 or denom_y == 0:
        return 0.0
    
    return round(numeratore  / (denom_x * denom_y), 4)

print("--- MATRICE DI CORRELAZIONE ---")
print(f"{'':15}", end="")
for n in nomi_var:
    print(f"{n:15}", end="")
print()
print("-" * 72)

for n1 in nomi_var:
    print(f"{n1:15}", end="")
    for n2 in nomi_var:
        r = calcola_correlazione(variabili[n1], variabili[n2])
        print(f"{r:15.3f}", end="")
    print()

print("\n" + "="*60 + "\n")



SOGLIA_CORRELAZIONE = 0.95
print(f"--- ANALISI RIDONDANZA (Soglia: {SOGLIA_CORRELAZIONE}) ---")

da_eliminare = set()
for i, n1 in enumerate(nomi_var):
    for j, n2 in enumerate(nomi_var):
        if i < j:
            r = calcola_correlazione(variabili[n1], variabili[n2])
            if abs(r) > SOGLIA_CORRELAZIONE:
                print(f"    {n1:12} '<->' {n2:12}  : r = {r:.3f} --> candidata eliminazione: {n2}")
                da_eliminare.add(n2)

variabili_mantenute = [n for n in nomi_var if n not in da_eliminare]
print(f"\nVariabili eliminate: {sorted(da_eliminare)}")
print(f"Variabili mantenute:  {variabili_mantenute}")
```
