```python
# Passo 2: Calcola media di eta escludendo None e valori > 100
nomi = ["Luca", "Sara", "Marco", "Paolo", "Anna"]
eta = [25, None, 35, 310, 29]
stipendio = [2500, 2800, None, 3200, 2900]
reparto = ["IT", "HR", "Marketing", "IT", "HR"]

def calcola_media_eta(lista_eta):
    """
    Calcola la media di eta escludendo None e valori > 100,
    e sostituisce questi valori con la media calcolata.
    """
    # Filtra i valori validi
    eta_validi = [e for e in lista_eta if e is not None and e <= 100]
    
    # Calcola la media
    media_eta = sum(eta_validi) / len(eta_validi)
    print(f"Media eta (escludendo None e > 100): {media_eta}")
    
    # Sostituisci i valori None e > 100 con la media
    eta_corretta = [media_eta if (e is None or e > 100) else e for e in lista_eta]
    print(f"Eta dopo la sostituzione: {eta_corretta}")
    
    return eta_corretta

# Applica la funzione
eta = calcola_media_eta(eta)
```

## Passo 3: Cacola la mediana si stipendio

```python
# Passo 3: Calcola mediana di stipendio escludendo None

nomi = ["Luca", "Sara", "Marco", "Paolo", "Anna"]
eta = [25, None, 35, 310, 29]
stipendio = [2500, 2800, None, 3200, 2900]
reparto = ["IT", "HR", "Marketing", "IT", "HR"]

def calcola_mediana_stipendio(lista_stipendio):
    """
    Calcola la mediana di stipendio escludendo None
    e sostituisce il valore mancante con la mediana
    """
    
    stipendio_valido = [s for s in lista_stipendio if s is not None]

    # Ordino i valori
    stipendio_valido.sort()

    # calcolo della mediana
    n = len(stipendio_valido)
    if n % 2 == 1:  # dispari
        mediana = stipendio_valido[n // 2]
    else: # pari
        mediana = (stipendio_valido[n // 2 - 1] + stipendio_valido[n // 2]) / 2

    print(f"La mediana dello stipendio escludendo il None è: {mediana}")
    
    # Sostituisci None con la mediana
    stipendio_corretto = [mediana if s is None else s for s in lista_stipendio]
    print(f"Stipendio dopo la sostituzione: {stipendio_corretto}")
    
    return stipendio_corretto

# Applica la funzione
stipendio = calcola_mediana_stipendio(stipendio)
```

## Passo 4: Applica l'encoding

```python
# 1. Dataset originale
nomi = ["Luca", "Sara", "Marco", "Paolo", "Anna"]
eta = [25, None, 35, 310, 29]
stipendio = [2500, 2800, None, 3200, 2900]
reparto = ["IT", "HR", "Marketing", "IT", "HR"]

# 2. Schema di codifica richiesto
schema_reparto = {
    'HR': 0, 
    'IT': 1, 
    'Marketing': 2
}

# 3. Applicazione dell'Encoding Ordinale (List Comprehension)
reparto_codificato = [schema_reparto[val] for val in reparto]

# Output del risultato specifico richiesto
print("Mappatura della colonna reparto:")
print(f"Originale:  {reparto}")
print(f"Codificato: {reparto_codificato}")
```

```python
# Passo 5: Stampa il dataset finale con tutti i valori corretti

# 1. Dataset originale
nomi = ["Luca", "Sara", "Marco", "Paolo", "Anna"]
eta = [25, None, 35, 310, 29]
stipendio = [2500, 2800, None, 3200, 2900]
reparto = ["IT", "HR", "Marketing", "IT", "HR"]

# 2. Calcola media di eta (escludendo None e valori > 100)
eta_validi = [e for e in eta if e is not None and e <= 100]
media_eta = sum(eta_validi) / len(eta_validi)
eta = [media_eta if (e is None or e > 100) else e for e in eta]

# 3. Calcola mediana di stipendio (escludendo None)
stipendio_valido = [s for s in stipendio if s is not None]
stipendio_valido.sort()
n = len(stipendio_valido)
mediana_stipendio = stipendio_valido[n // 2] if n % 2 == 1 else (stipendio_valido[n // 2 - 1] + stipendio_valido[n // 2]) / 2
stipendio = [mediana_stipendio if s is None else s for s in stipendio]

# 4. Applica encoding ordinale al reparto
schema_reparto = {'HR': 0, 'IT': 1, 'Marketing': 2}
reparto_codificato = [schema_reparto[val] for val in reparto]

# 5. Stampa il dataset finale
print("=" * 80)
print("DATASET FINALE CON TUTTI I VALORI CORRETTI")
print("=" * 80)
print()
print(f"Nomi:                  {nomi}")
print(f"Eta (corretta):        {eta}")
print(f"Stipendio (corretto):  {stipendio}")
print(f"Reparto:               {reparto}")
print(f"Reparto (codificato):  {reparto_codificato}")
print()
print("=" * 80)
print("TABELLA RIEPILOGATIVA")
print("=" * 80)
for i in range(len(nomi)):
    print(f"{nomi[i]:<10} | Eta: {eta[i]:<20} | Stipendio: {stipendio[i]:<8} | Reparto: {reparto[i]:<12} | Cod: {reparto_codificato[i]}")
```
