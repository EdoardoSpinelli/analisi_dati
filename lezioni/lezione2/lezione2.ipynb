```python
import pandas as pd

gruppo_A = [1900, 2000, 2100, 2200, 2300]
gruppo_B = [500, 1000, 2100, 3200, 4500]

s_A = pd.Series(gruppo_A)
s_B = pd.Series(gruppo_B)

print("Gruppo A")
print("Media: ", s_A.mean())
print("Mediana: ", s_A.median())
print("Std: ", s_A.std())
print("Min: ", s_A.min())
print("Max: ", s_A.max())
print("Q1: ", s_A.quantile(0.25))
print("Q3: ", s_A.quantile(0.75))
print("IQR: ", s_A.quantile(0.75) - s_A.quantile(0.25))




print("\n Gruppo B")
print("Media: ", s_B.mean())
print("Mediana: ", s_B.median())
print("Std: ", s_B.std())
print("Min: ", s_B.min())
print("Max: ", s_B.max())
print("Q1: ", s_B.quantile(0.25))       # corrisponde al 25%
print("Q2: ", s_B.quantile(0.75))       # Corrisponde alla mediana
print("IQR: ", s_B.quantile(0.75) - s_B.quantile(0.25))     # la larghezza della scatola centrale z
```

```python
eta_dipendenti = [28, -5, 35, 150, 29, 0, 42, "N/A", 31]

def analizza_eta(lista):
    valori_corretti = []
    valori_errati = []
    valori_mancanti = []
    
    for x in lista:
        if x == "N/A" or x is None:
            valori_mancanti.append(x)
        elif not isinstance(x, (int, float)):
            valori_errati.append(x)
        elif x <= 15 or x > 90:  # Limiti logici aziendali
            valori_errati.append(x)
        else:
            valori_corretti.append(x)
            
    return valori_corretti, valori_errati, valori_mancanti

corretti, errati, mancanti = analizza_eta(eta_dipendenti)
print("Corretti:", corretti) # [28, 35, 29, 42, 31]
print("Errati:", errati)     # [-5, 150, 0]
print("Mancanti:", mancanti) # ['N/A']
```

```python
stipendi = [2100, None, 1800, None, 2400, 1950, 15000]


def calcola_media(lista):
    valori_validi = [x for x in lista if x is not None]
    if not valori_validi:
        return 0
    return sum(valori_validi) / len(valori_validi)


def calcola_mediana(lista):
    valori_validi = [x for x in lista if x is not None]
    if not valori_validi:
        return 0
    
    valori_ordinati = sorted(valori_validi)
    n = len(valori_ordinati)
    centro = n // 2
    
    if n % 2 != 0:
        return valori_ordinati[centro]
    else:
        return (valori_ordinati[centro - 1] + valori_ordinati[centro]) / 2


def sostituisci_mancanti_con_mediana(lista):
    
    mediana_valore = calcola_mediana(lista)
    
    # Creiamo una nuova lista sostituendo None con la mediana calcolata
    lista_pulita = [x if x is not None else mediana_valore for x in lista]
    return lista_pulita, mediana_valore

media_con_outlier = calcola_media(stipendi)
mediana_con_outlier = calcola_mediana(stipendi)

print("--- ANALISI DATI ORIGINALI ---")
print(f"Stipendi validi ordinati: {sorted([x for x in stipendi if x is not None])}")
print(f"Media degli stipendi (con outlier): {media_con_outlier:.2f} €")
print(f"Mediana degli stipendi (con outlier): {mediana_con_outlier:.2f} €")
print("-" * 30)

stipendi_imputati, med_usata = sostituisci_mancanti_con_mediana(stipendi)

print("\n--- SOSTITUZIONE VALORI MANCANTI ---")
print(f"Valore utilizzato per la sostituzione (Mediana): {med_usata} €")
print(f"Nuovo dataset con valori sostituiti: {stipendi_imputati}")
```
