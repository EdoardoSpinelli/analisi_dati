```python
stipendi_con_outlier = [2100, None, 1800, None, 2400, 1950, 15000]

def calcola_media(lista):
    valori_validi = [x for x in lista if x is not None]
    if not valori_validi:
        return 0
    return sum(valori_validi) / len(valori_validi)

media = calcola_media(stipendi_con_outlier)
print(f" La media è {media}")


def calcola_mediana(lista):
    valori_validi = [x for x in lista if x is not None]
    if not valori_validi:
        return 0

    valori_ordinati = sorted(valori_validi)
    n = len(valori_ordinati)
    centro = n // 2

    if n % 2 != 0:
        return valori_ordinati[centro]
    else:
            retrun (valori_ordinati[centro - 1] + valori_ordinati[centro]) / 2
    

mediana = calcola_mediana(stipendi_con_outlier)
print(f"La mediana è: {mediana}")
    
```

```python
citta = ["Milano", "Roma", None, "Milano", "Torino", None, "Milano", "Roma"]

def calcola_moda(lista):
    # 1. Puliamo la lista dai valori None
    valori_validi = [x for x in lista if x is not None]
    
    if not valori_validi:
        return None
        
    # 2. Creiamo un dizionario per contare le frequenze
    conteggio = {}
    for elemento in valori_validi:
        if elemento in conteggio:
            conteggio[elemento] += 1
        else:
            conteggio[elemento] = 1
            
    # 3. Troviamo la chiave con il valore di conteggio più alto
    moda_piu_frequente = None
    massimo_voti = 0
    
    for chiave, valore in conteggio.items():
        if valore > massimo_voti:
            massimo_voti = valore
            moda_piu_frequente = chiave
            
    return moda_piu_frequente

# Eseguiamo la funzione
moda = calcola_moda(citta)
print(f"La moda è: {moda}")
# Output: Milano
```
