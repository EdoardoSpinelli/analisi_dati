```python
import numpy as np
import pandas as pd

# Array NumPy forniti nel testo
prezzi_array = np.array([12.5, 30.0, 8.75, 45.0, 22.0, 15.5, 60.0, 9.99])
quantita_array = np.array([10, 4, 25, 2, 8, 15, 3, 20])

# Dizionario dati fornito nel testo
dati = {
    "venditore": ["Alice", "Bob", "Alice", "Carlo", "Bob", "Alice", "Carlo", "Bob", "Carlo", "Alice", "Bob", "Carlo"],
    "regione": ["Nord", "Sud", "Nord", "Centro", "Sud", "Nord", "Centro", "Sud", "Centro", "Nord", "Sud", "Centro"],
    "categoria": ["Elettronica", "Abbigliamento", "Elettronica", "Alimentari", "Elettronica", "Abbigliamento", "Elettronica", "Alimentari", "Abbigliamento", "Alimentari", "Elettronica", "Abbigliamento"],
    "quantita": [5, 12, 3, 20, 8, 15, 6, 25, 10, 18, 4, 9],
    "prezzo": [200.0, 45.0, 200.0, 8.5, 200.0, 45.0, 200.0, 8.5, 45.0, 8.5, 200.0, 45.0],
    "mese": ["Gen", "Gen", "Feb", "Gen", "Feb", "Mar", "Feb", "Mar", "Mar", "Gen", "Mar", "Feb"]
}
```

# Obiettivo 1: Creare e manipolare array NumPy

```python
# Operazione vettoriale: Calcolo del ricavo totale per elemento
ricavo_per_elemento = prezzi_array * quantita_array
print("Ricavo per ogni elemento dell'array:", ricavo_per_elemento)

# Statistiche di base
print("Quantità totale venduta (NumPy):", np.sum(quantita_array))
print("Prezzo medio (NumPy):", np.mean(prezzi_array))
print("Prezzo massimo (NumPy):", np.max(prezzi_array))
```

# Obiettivo 2: Costruire un DataFrame Pandas

```python
df = pd.DataFrame(dati)
print("DataFrame creato con successo:\n", df.head())
```

## Obiettivo 3 & 4: 
### Selezionare, filtrare dati con .loc[], .iloc[] e condizioni booleane
### Utilizziamo gli indicizzatori e i criteri logici (& per AND, | per OR, ~ per NOT).

```python
# Selezione con .iloc (le prime 5 righe e le prime 3 colonne)
print("\n--- Selezione con .iloc ---")
print(df.iloc[0:5, 0:3])

# Filtro condizionale con .loc: Vendite in regione "Nord" CON quantità maggiore di 10
print("\n--- Filtro .loc (Nord AND quantita > 10) ---")
condizione_nord_alta_qty = (df['regione'] == 'Nord') & (df['quantita'] > 10)
print(df.loc[condizione_nord_alta_qty])

# Filtro condizionale: Categoria "Elettronica" OPPURE regione "Centro"
print("\n--- Filtro (Elettronica OR Centro) ---")
condizione_or = (df['categoria'] == 'Elettronica') | (df['regione'] == 'Centro')
print(df.loc[condizione_or].head())
```

## Obiettivo 5: Aggiungere colonne calcolate a un DataFrame

```python
df['totale_vendite'] = df['quantita'] * df['prezzo']
print("\n--- DataFrame con colonna calcolata 'totale_vendite' ---")
print(df.head())
```

## Obiettivo 6: Usare groupby() per aggregare i dati

```python
print("\n--- Aggregazione con groupby() per Categoria ---")
aggregato_categoria = df.groupby('categoria')[['quantita', 'totale_vendite']].sum()
print(aggregato_categoria)
```

## Obiettivo 7: Costruire una Pivot Table

```python
print("\n--- Pivot Table (Mese vs Regione) ---")
pivot_report = pd.pivot_table(
    df, 
    values='totale_vendite', 
    index='regione', 
    columns='mese', 
    aggfunc='sum', 
    fill_value=0
)
print(pivot_report)
```
