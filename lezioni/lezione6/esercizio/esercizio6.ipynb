```python
import numpy as np

altezze = np.array([170, 165, 180, 155, 175, 160, 190, 168, 172, 158])

pesi = np.array([70, 58, 85, 52, 78, 65, 95, 72, 68, 55])

pressione = np.array([120, 118, 125, 115, 122, 119, 128, 121, 195, 116])

dati = np.column_stack([altezze, pesi, pressione])


print("--- PUNTO 1 ---")
print("Matrice dati (10x3):\n", dati)
print("Shape:", dati.shape)
print("Dtype:", dati.dtype)
print("\n" + "-"*40 + "\n")


# Standardizzazione con Z-score

dati_standardizzati = np.zeros_like(dati, dtype=float)

numero_colonne = dati.shape[1]
for i in range(numero_colonne):
    colonna = dati[:, 1]
    media_colonna = np.mean(colonna)
    deviazione_standard_colonna = np.std(colonna)

    dati_standardizzati[:, i] = (colonna - media_colonna) / deviazione_standard_colonna

media_verificata =  np.mean(dati_standardizzati, axis=0)
deviazione_standard_verifca = np.std(dati_standardizzati, axis=0)


print("--- PUNTO 2 ---")
print("Matrice standardizzata:\n", dati_standardizzati)
print("Verifica Medie (circa 0):", np.round(media_verificata, 2))
print("Verifica Deviazioni Standard (circa 1):", np.round(deviazione_standard_verifca, 2))
print("\n" + "-"*40 + "\n")



# --- 3. Rileva outlier: IQR ---

Q1 = np.percentile(dati, 25, axis=0)
Q3 = np.percentile(dati, 75, axis=0)
IQR = Q3 - Q1

limite_inferiore = Q1 - 1.5 * IQR
limite_superiore = Q3 + 1.5 * IQR

# Individua i pazienti che hanno almeno una misurazione fuori dai limiti
outliers = np.any((dati < limite_inferiore) | (dati > limite_superiore), axis=1)

print("--- PUNTO 3 ---")
print("Q1 per colonna:", Q1)
print("Q3 per colonna:", Q3)
print("IQR per colonna:", IQR)
print("Pazienti outlier (True indica che è un outlier):", outliers)
print("\n" + "-"*40 + "\n")



# --- 4. Calcola la matrice di covarianza ---

matrice_cov = np.cov(dati_standardizzati, rowvar=False)

print("--- PUNTO 4 ---")
print("Matrice di covarianza (sui dati standardizzati):\n", matrice_cov)
print("\n" + "-"*40 + "\n")
```
