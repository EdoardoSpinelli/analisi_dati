### 1. Costruzione ed ispezione
Crea il DataFrame con i dati sopra. Verifica la struttura: quante righe e colonne ha, i tipi di dato di ogni colonna e stampa le prime 5 righe. Calcola le statistiche descrittive sulle colonne numeriche.

```python
import pandas as pd

# Ripristiniamo il DataFrame iniziale
data = {
    "Data": ["2023-01-15", "2023-02-20", "2023-03-10", "2023-04-05", "2023-05-18", "2023-06-22", "2023-07-14", "2023-08-09", "2023-09-30", "2023-10-11", "2023-11-25", "2023-12-20"],
    "Città": ["Milano", "Roma", "Milano", "Torino", "Roma", "Milano", "Torino", "Roma", "Milano", "Torino", "Roma", "Milano"],
    "Categoria": ["Invernale", "Invernale", "Primavera", "Primavera", "Primavera", "Estiva", "Estiva", "Estiva", "Autunno", "Autunno", "Autunno", "Invernale"],
    "Vendite (€)": [12400, 9800, 7200, 6100, 8300, 15200, 13700, 14100, 9500, 7800, 8600, 18900],
    "N. Transazioni": [310, 245, 180, 152, 208, 380, 342, 352, 237, 195, 215, 472]
}
df = pd.DataFrame(data)

# Verifica la struttura (righe e colonne)
print(f"Dimensioni del DataFrame: {df.shape} (Righe, Colonne)\n")

# Tipi di dato di ogni colonna
print("Tipi di dato delle colonne:")
print(df.dtypes)
print("\n" + "-"*40 + "\n")

# Stampa le prime 5 righe
print("Prime 5 righe:")
display(df.head(5))
print("\n" + "-"*40 + "\n")

# Statistiche descrittive sulle colonne numeriche
print("Statistiche descrittive:")
display(df.describe())
```

### 2. Selezione, filtraggio e colonne calcolate
* Seleziona solo le colonne data, città e vendite.
* Filtra le righe con vendite superiori a 10.000€.
* Filtra le righe di Milano oppure Roma con vendite superiori a 8.000€.
* Aggiungi una colonna `valore_medio_transazione` calcolata come vendite / n_transazioni (arrotondata a 2 decimali).
* Aggiungi una colonna `performance` usando `apply()`: "Alta" se vendite > 12.000, "Media" se vendite > 8.000, "Bassa" altrimenti.

```python
# Seleziona solo le colonne data, città e vendite
df_selezione = df[["Data", "Città", "Vendite (€)"]]

# Filtra le righe con vendite superiori a 10.000€
df_filtro_10k = df[df["Vendite (€)"] > 10000]

# Filtra le righe di Milano oppure Roma con vendite superiori a 8.000€
df_filtro_citta_8k = df[df["Città"].isin(["Milano", "Roma"]) & (df["Vendite (€)"] > 8000)]

# Aggiungi colonna valore_medio_transazione (arrotondata a 2 decimali)
df["valore_medio_transazione"] = (df["Vendite (€)"] / df["N. Transazioni"]).round(2)

# Aggiungi colonna performance usando apply()
def valuta_performance(row):
    if row["Vendite (€)"] > 12000:
        return "Alta"
    elif row["Vendite (€)"] > 8000:
        return "Media"
    else:
        return "Bassa"

df["performance"] = df.apply(valuta_performance, axis=1)

# Visualizziamo il DataFrame aggiornato
display(df)
```

### 3. Aggregazione e Pivot Table
* Calcola per ogni città: totale vendite, media vendite e numero di mesi registrati.
* Calcola per ogni categoria: media vendite e deviazione standard.
* Costruisci una Pivot Table con città sulle righe, categoria sulle colonne e la somma delle vendite nelle celle. Usa `fill_value=0`.

```python
# Per ogni città: totale vendite, media vendite e numero di mesi registrati
# (Nota: assumiamo che ogni riga rappresenti un mese unico registrato)
agg_citta = df.groupby("Città").agg(
    Totale_Vendite=("Vendite (€)", "sum"),
    Media_Vendite=("Vendite (€)", "mean"),
    Mesi_Registrati=("Data", "count")
)
print("Aggregazione per Città:")
display(agg_citta)

# Per ogni categoria: media vendite e deviazione standard
agg_categoria = df.groupby("Categoria").agg(
    Media_Vendite=("Vendite (€)", "mean"),
    Deviazione_Standard=("Vendite (€)", "std")
)
print("\nAggregazione per Categoria:")
display(agg_categoria)

# Pivot Table: città sulle righe, categoria sulle colonne, somma delle vendite
pivot_vendite = df.pivot_table(
    values="Vendite (€)", 
    index="Città", 
    columns="Categoria", 
    aggfunc="sum", 
    fill_value=0
)
print("\nPivot Table (Somma Vendite):")
display(pivot_vendite)
```

### 4. Analisi temporale
* Converti la colonna data in formato datetime.
* Estrai mese, nome del giorno della settimana e trimestre in tre nuove colonne.
* Calcola il totale delle vendite per trimestre.
* Calcola la media delle vendite per mese e individua il mese con il picco massimo.

```python
# Converti la colonna data in formato datetime
df["Data"] = pd.to_datetime(df["Data"])

# Estrai mese, nome del giorno della settimana e trimestre in tre nuove colonne
df["Mese"] = df["Data"].dt.month
df["Giorno_Settimana"] = df["Data"].dt.day_name()  # In inglese (es. Monday) o dt.day_of_week per l'indice
df["Trimestre"] = df["Data"].dt.to_period("Q")

print("DataFrame con colonne temporali estratte:")
display(df.head(3))

# Calcola il totale delle vendite per trimestre
vendite_trimestre = df.groupby("Trimestre")["Vendite (€)"].sum().reset_index()
print("\nTotale vendite per Trimestre:")
display(vendite_trimestre)

# Calcola la media delle vendite per mese e individua il mese con il picco massimo
# (Nel dataset ogni mese compare una sola volta, quindi la media coincide con il valore stesso)
media_mensile = df.groupby("Mese")["Vendite (€)"].mean()
mese_picco = media_mensile.idxmax()
valore_picco = media_mensile.max()

print(f"\nLa media delle vendite per mese è:\n{media_mensile.to_string()}")
print(f"\nIl mese con il picco massimo di vendite è il mese {mese_picco} con un valore di {valore_picco}€.")
```
