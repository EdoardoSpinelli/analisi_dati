## Normalizzazione

```python
# Blocco 1: Definizione delle liste dati
eta = [22, 28, 35, 45, 52, 30, 40, 25]
stipendio = [1800, 2200, 3500, 5000, 7000, 2800, 4500, 2000]
esperienza = [1, 3, 8, 18, 25, 5, 14, 2]

# Blocco 2: Funzione di normalizzazione Min-Max
def normalizza_minmax(lista):
    minimo = min(lista)
    massimo = max(lista)
    rango = massimo - minimo
    
    if rango == 0:
        return [0.5 for _ in lista]
        
    risultato = []
    for x in lista:
        x_norm = (x - minimo) / rango
        risultato.append(round(x_norm, 4))
        
    return risultato

# Blocco 3: Esempio pratico di calcolo passo-passo con formattazione tabellare
eta = [22, 28, 35, 45, 52, 30, 40, 25]

minimo = min(eta)
massimo = max(eta)
rango = massimo - minimo

print(f"Età minima  : {minimo}")
print(f"Età massima : {massimo}")
print(f"Rango       : {rango}")

print() # Riga vuota per separare l'output
print(f"{'Età orig':10} {'Calcolo':30} {'Risultato':10}")
print("-" * 52) # I num 10 e 30: servono a dare una larghezza fissa alle colonne, così la tabella si allinea bene.

for x in eta:
    risultato = (x - minimo) / rango
    calcolo = f"({x} - {minimo}) / {rango}"
    print(f"{x:10} {calcolo:30} {risultato:.4f}")
```

## Standardizzazione

```python
def calcola_media(lista):
    return sum(lista) / len(lista)

def calcola_deviazione_standard(lista):
    media = calcola_media(lista)
    scarti_quadrati = [(x - media) ** 2 for x in lista]
    varianza = sum(scarti_quadrati) / (len(lista) - 1)
    return varianza ** 0.5

def standardizza_zscore(lista):
    media = calcola_media(lista)
    sigma = calcola_deviazione_standard(lista)
    
    if sigma == 0:
        return [0.0 for _ in lista]
        
    risultato = []
    for x in lista:
        z = (x - media) / sigma
        risultato.append(round(z, 4))
        
    return risultato

# Sezione di esempio pratico
stipendio = [1800, 2200, 3500, 5000, 7000, 2800, 4500, 2000]

media = calcola_media(stipendio)
sigma = calcola_deviazione_standard(stipendio)

print(f"Media stipendio        : {media:.2f} €")
print(f"Deviazione standard    : {sigma:.2f} €")
print()
print(f"{'Stipendio':12} {'Calcolo':35} {'Z-score':10}")
print("-" * 60)

for x in stipendio:
    z = (x - media) / sigma
    calcolo = f"({x} - {media:.0f}) / {sigma:.0f}"
    interpretazione = "sopra media" if z > 0 else "sotto media"
    print(f"{x:12} {calcolo:35} {z:+.4f}  ({interpretazione})")
```

```python
def one_hot_encoding(lista, prefisso="cat"):
    categorie = sorted(set(lista))
    
    encoded = {}
    for cat in categorie:
        nome_colonna = f"{prefisso}_{cat}"
        encoded[nome_colonna] = []
        for valore in lista:
            if valore == cat:
                encoded[nome_colonna].append(1)
            else:
                encoded[nome_colonna].append(0)
                
    return encoded, categorie

nomi = ["Anna", "Luca", "Sara", "Marco", "Giulia"]
citta = ["Roma", "Milano", "Torino", "Roma", "Milano"]

encoded, categorie = one_hot_encoding(citta, prefisso="città")

intestazione = f"{'Nome':8}"
for cat in categorie:
    intestazione += f" {'città_'+cat:15}"
print(intestazione)
print("-" * (8 + 15 * len(categorie)))

for i, nome in enumerate(nomi):
    riga = f"{nome:8}"
    for cat in categorie:
        val = encoded[f"città_{cat}"][i]
        riga += f" {val:15}"
    print(riga)
```
